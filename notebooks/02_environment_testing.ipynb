```python
from inspect import signature
import sklearn.ensemble

cls = sklearn.ensemble.GradientBoostingClassifier
sig = signature(cls)
for name, param in sig.parameters.items():
    print(name)
    #'\t', param.default, '\t', type(param.default).__name__,)
    for command in dir(param):
        result = eval(f"param.{command}")
        # Only call if it's callable and not a string
        if callable(result):
            try:
                result = result()
            except Exception:
                result = '«FALHA» ' + str(result)  # Fallback to string representation if calling fails
                pass
        print(command, '\t', result)
    print('---')
```

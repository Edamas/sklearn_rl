```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import make_classification
import numpy as np  # <--- necessário

# gerar um dataset pequeno e compatível
X, y = make_classification(
    n_samples=200, 
    n_features=4,
    n_informative=2,
    n_redundant=0,
    random_state=0,
    shuffle=False
)

clf = AdaBoostClassifier(n_estimators=50, random_state=0)
clf.fit(X, y)

# predição com entrada artificial
pred = clf.predict([[0, 0, 0, 0]])
print("Predição:", pred)        # ex.: [1]

# acurácia no próprio conjunto de treino
score = clf.score(X, y)
print("Acurácia:", score)
```

```python
import sklearn
import pandas as pd
import inspect
from sklearn.utils import all_estimators

def clean_doc(obj):
    try:
        doc = getattr(obj, "__doc__", None)
        if not doc:
            return "Sem descrição"
        lines = [l.strip() for l in doc.splitlines() if l.strip()]
        if not lines:
            return "Sem descrição"
        first_line = lines[0]
        skip_chars = ("=", ":", "`", ".", "@")
        if first_line.startswith(skip_chars) and len(lines) > 1:
            return lines[1]
        return first_line
    except Exception as e:
        return f"Erro ao obter doc: {e}"

def get_params(cls):
    try:
        sig = inspect.signature(cls.__init__)
        return ",".join(sig.parameters.keys())
    except Exception:
        return ""

records = []

# --- 1) Coleta de módulos e métodos (até nível 4) ---
for mod1_name in dir(sklearn):
    if mod1_name.startswith("_"):
        continue
    try:
        mod1_obj = getattr(sklearn, mod1_name)
        records.append({
            "mod1": mod1_name, "mod2": "", "mod3": "", "mod4": "",
            "mod_doc": clean_doc(mod1_obj), "estimator_type": "", "parameters": ""
        })
        for mod2_name in dir(mod1_obj):
            if mod2_name.startswith("_"):
                continue
            try:
                mod2_obj = getattr(mod1_obj, mod2_name)
                records.append({
                    "mod1": mod1_name, "mod2": mod2_name, "mod3": "", "mod4": "",
                    "mod_doc": clean_doc(mod2_obj), "estimator_type": "", "parameters": ""
                })
                for mod3_name in dir(mod2_obj):
                    if mod3_name.startswith("_"):
                        continue
                    try:
                        mod3_obj = getattr(mod2_obj, mod3_name)
                        records.append({
                            "mod1": mod1_name, "mod2": mod2_name, "mod3": mod3_name, "mod4": "",
                            "mod_doc": clean_doc(mod3_obj), "estimator_type": "", "parameters": ""
                        })
                        if inspect.isclass(mod3_obj):
                            for mod4_name in dir(mod3_obj):
                                if mod4_name.startswith("_"):
                                    continue
                                try:
                                    mod4_obj = getattr(mod3_obj, mod4_name)
                                    records.append({
                                        "mod1": mod1_name, "mod2": mod2_name, "mod3": mod3_name, "mod4": mod4_name,
                                        "mod_doc": clean_doc(mod4_obj), "estimator_type": "", "parameters": ""
                                    })
                                except Exception:
                                    continue
                    except Exception:
                        continue
            except Exception:
                continue
    except Exception:
        continue

# --- 2) Coleta de todos os estimators ---
type_filters = ["classifier", "regressor", "cluster", "transformer"]

for t in type_filters:
    for name, cls in all_estimators(type_filter=t):
        # caminho real do módulo + classe
        module_path = cls.__module__.split(".")[1:]  # remove 'sklearn'
        qual_path = cls.__qualname__.split(".")
        full_path = module_path + qual_path
        # garante exatamente 4 níveis
        while len(full_path) < 4:
            full_path.append("")
        records.append({
            "mod1": full_path[0] if len(full_path) > 0 else "",
            "mod2": full_path[1] if len(full_path) > 1 else "",
            "mod3": full_path[2] if len(full_path) > 2 else "",
            "mod4": full_path[3] if len(full_path) > 3 else "",
            "mod_doc": clean_doc(cls),
            "estimator_type": t,
            "parameters": get_params(cls)
        })

df = pd.DataFrame(records, columns=["mod1","mod2","mod3","mod4","mod_doc","estimator_type","parameters"])
df.to_csv("sklearn_modules_exploration.csv", index=False, sep='\t', encoding='utf-8')
df
```

```python
import pandas as pd
import inspect
from sklearn.utils import all_estimators
pd.set_option("display.max_rows", 50)

# Lista de tipos de estimators que queremos
type_filters = ["classifier", "regressor", "cluster", "transformer"]

# Armazenar todos os dados temporariamente
rows = []

for type_filter in type_filters:
    estimators = all_estimators(type_filter=type_filter)
    for name, cls in estimators:
        # Iterar atributos da classe para identificar métodos de instância
        method_names = []
        for attr in dir(cls):
            try:
                member = getattr(cls, attr)
                if (inspect.isfunction(member) or inspect.ismethod(member)) and not attr.endswith('_'):
                    method_names.append(attr)
            except Exception:
                continue
        rows.append({
            "estimator_name": name,
            "estimator_type": type_filter,
            "methods_list": ', '.join(sorted(method_names))
        })

# Criar DataFrame final
df_estimators = pd.DataFrame(rows)

# Visualizar
df_estimators
print(df_estimators)
```

```python
methods_list = {}
for estimator_name, estimator_type, estimator_methods in zip(df_estimators['estimator_name'], df_estimators['estimator_type'], df_estimators['methods_list']):
    methods = estimator_methods.split(', ')
    for method in methods:
        key = (estimator_type, method)
        methods_list[key] = methods_list.get(key, 0) + 1

# Ordenar por contagem decrescente
sorted_methods = sorted(methods_list.items(), key=lambda x: x[1], reverse=True)

for (estimator_type, method), count in sorted_methods:
    print(f"{estimator_type}\t{method}\t{count}")
```

```python
import sklearn
import inspect
import pkgutil

# Métodos privados realmente úteis
allowed_private = [
    '_validate_data', 
    '_more_tags', 
    '_check_fit_params', 
    '_get_tags'
]

def safe_doc(obj):
    """Retorna a primeira linha da docstring de forma segura."""
    doc = getattr(obj, '__doc__', '')
    if isinstance(doc, str) and doc:
        return doc.strip().splitlines()[0]
    return "Sem descrição"

def extract_sklearn_docs(include_private=False):
    """
    Extrai até 3 níveis de módulos/classes/métodos do sklearn.
    
    include_private: bool
        Se True, inclui todos os métodos privados.
    """
    results = []

    for importer, modname, ispkg in pkgutil.iter_modules(sklearn.__path__, prefix='sklearn.'):
        # ignora módulos totalmente privados
        if modname.split('.')[-1].startswith('_') and modname.split('.')[-1] not in allowed_private:
            continue

        try:
            mod1 = __import__(modname, fromlist=[''])
        except Exception:
            continue

        mod1_name = modname
        mod1_doc = safe_doc(mod1)
        results.append((mod1_name, mod1_doc, '', '', '', ''))

        for name2, obj2 in inspect.getmembers(mod1):
            # filtra nomes privados exceto os permitidos
            if (name2.startswith('_') or name2.startswith('__')) and name2 not in allowed_private:
                continue
            if inspect.ismodule(obj2) or inspect.isclass(obj2) or inspect.isfunction(obj2):
                mod2_name = name2
                mod2_doc = safe_doc(obj2)
                results.append((mod1_name, mod1_doc, mod2_name, mod2_doc, '', ''))

                # nível 3
                if inspect.isclass(obj2) or inspect.ismodule(obj2):
                    for name3, obj3 in inspect.getmembers(obj2):
                        if (name3.startswith('_') or name3.startswith('__')) and name3 not in allowed_private:
                            continue
                        if inspect.isfunction(obj3) or inspect.ismethod(obj3):
                            mod3_name = name3
                            mod3_doc = safe_doc(obj3)
                            results.append((mod1_name, mod1_doc, mod2_name, mod2_doc, mod3_name, mod3_doc))

    return results

# Exemplo de uso
docs = extract_sklearn_docs(include_private=False)
for r in docs:
    print("\t".join([str(c) for c in r]))
```
